```python
import pymongo
import pprint
from connect import uri
```

```python
# Update with your connection URI from the Atlas UI
# and create database object
mc = pymongo.MongoClient(uri('3.6'))
mflix = mc['mflix']
```

```python
# show first record as example
pprint.pprint(mflix.movies_initial.find_one())
```

```python
# find all movies released from 1983 onwards
filters = {'year': { '$gte': 1983 }}

for movie in mflix.movies_initial.find(filters):
    pprint.pprint(movie['title'])
```

```python
# find all movies between 1989 and 1999
filters = { 'year': {'$gte': 1989, '$lt': 2000} }

for movie in mflix.movies_initial.find(filters):
    pprint.pprint(movie['title'])
```

```python
# find all movies released in 1995, 2005, 2015
filters = { 'year': { '$in': [ 1995, 2005, 2015 ] } }
for movie in mflix.movies_initial.find(filters):
    pprint.pprint(movie['title'])
```

```python
# find all movies except the ones which are of adult genre
filters = { 'year': { '$in': [ 1995, 2005, 2015 ] }, 
                     'genre': { '$not' : {'$eq': 'Adult'} } }

for movie in mflix.movies_initial.find(filters):
    pprint.pprint(movie['title'])
```
